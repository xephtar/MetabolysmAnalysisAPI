```python
from textblob import TextBlob
import requests
import nltk
nltk.download('punkt')
import spacy
from spacy import displacy
nlp = spacy.load("en_ner_bc5cdr_md")
nlp_bio = spacy.load("en_ner_bionlp13cg_md")
```

```python
url = "http://localhost:8000/api/articles/?search=alzheimer"
response = requests.request("GET", url)
data = response.json()['results'][7]['abstract_text']
```

```python
doc = nlp(data)
for ent in doc.ents:
    if ent.label_ in ["DISEASE"]:
        print(ent.text, ent.start_char, ent.end_char, ent.label_)
displacy.render(doc,jupyter=True,style='ent')
```

```python
doc_bio = nlp_bio(data)
for ent in doc_bio.ents:
    print(ent.text, ent.start_char, ent.end_char, ent.label_)
displacy.render(doc_bio,jupyter=True,style='ent')
```
